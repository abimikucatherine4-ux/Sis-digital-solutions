```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/content/diabetes (2).csv")

# 1. How many patients have diabetes?
diabetic_patients = df[df["Outcome"] == 1].shape[0]
print("1. Patients with diabetes:", diabetic_patients)
print("This makes sense because Outcome = 1 represents patients diagnosed with diabetes.\n")

# 2. What is the average glucose level?
average_glucose = df["Glucose"].mean()
print("2. Average glucose level:", average_glucose)
print("This makes sense because the mean gives the overall average glucose reading of all patients.\n")

# 3. Show only female patients above 40
female_above_40 = df[(df["Age"] > 40)]
print("3. Female patients above 40:")
print(female_above_40.head())
print("This makes sense because filtering selects only patients who match the age condition.\n")

# 4. Group by Outcome and show average BMI
average_bmi = df.groupby("Outcome")["BMI"].mean()
print("4. Average BMI grouped by Outcome:")
print(average_bmi)
print("This makes sense because grouping compares BMI averages between diabetic and non-diabetic patients.\n")

# 5. Drop rows with missing or zero glucose values
original_rows = df.shape[0]
cleaned_df = df[df["Glucose"] > 0].dropna(subset=["Glucose"])
removed_rows = original_rows - cleaned_df.shape[0]

print("5. Rows removed:", removed_rows)
print("This makes sense because rows with invalid or missing glucose values are removed for cleaner analysis.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/diabetes (2).csv")

# 1. Histogram
plt.figure(figsize=(6,4))
plt.hist(df["Glucose"], bins=10)
plt.title("Histogram of Glucose Levels")
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
plt.show()

# 2. Scatter Plot
plt.figure(figsize=(6,4))
plt.scatter(df["Age"], df["BMI"])
plt.title("Scatter Plot of Age vs BMI")
plt.xlabel("Age")
plt.ylabel("BMI")
plt.show()

# 3. Bar Chart
outcome_counts = df["Outcome"].value_counts()

plt.figure(figsize=(6,4))
plt.bar(outcome_counts.index, outcome_counts.values)
plt.title("Bar Chart of Diabetes Outcomes")
plt.xlabel("Outcome")
plt.ylabel("Number of Patients")
plt.show()

# 4. Box Plot
plt.figure(figsize=(6,4))
plt.boxplot(df["Glucose"])
plt.title("Box Plot of Glucose Levels")
plt.xlabel("Glucose")
plt.ylabel("Values")
plt.show()

# 5. Line Chart
plt.figure(figsize=(6,4))
plt.plot(df["Age"])
plt.title("Line Chart of Patient Ages")
plt.xlabel("Patient Index")
plt.ylabel("Age")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/content/diabetes (2).csv")

# Apply seaborn styling
sns.set_style("darkgrid")

# 1. Histogram
plt.figure(figsize=(6,4))
sns.histplot(df["Glucose"], bins=10)
plt.title("Histogram of Glucose Levels")
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
plt.show()

# 2. Scatter Plot
plt.figure(figsize=(6,4))
sns.scatterplot(x=df["Age"], y=df["BMI"])
plt.title("Scatter Plot of Age vs BMI")
plt.xlabel("Age")
plt.ylabel("BMI")
plt.show()

# 3. Bar Chart
plt.figure(figsize=(6,4))
sns.countplot(x=df["Outcome"])
plt.title("Bar Chart of Diabetes Outcomes")
plt.xlabel("Outcome")
plt.ylabel("Number of Patients")
plt.show()

# 4. Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(y=df["Glucose"])
plt.title("Box Plot of Glucose Levels")
plt.xlabel("Glucose")
plt.ylabel("Values")
plt.show()

# 5. Line Chart
plt.figure(figsize=(6,4))
sns.lineplot(x=df.index, y=df["Age"])
plt.title("Line Chart of Patient Ages")
plt.xlabel("Patient Index")
plt.ylabel("Age")
plt.show()
```

I prefer the seaborn version because;
1.The  charts look cleaner and more professional.
2.It adds better styling and  makes data easier to read with the grids and spacing.
